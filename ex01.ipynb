Import Python libraries


```python
import pandas as pd
import plotly.express as px
```

Read the CSV file

```python
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/station_Info.csv")
```

Extract Latitude and Longitude from the 'Location' column

```python
df[['Longitude', 'Latitude']] = df['Location'].str.extract(r'Point\((.*?) (.*?)\)')
```

Convert the extracted Latitude and Longitude to numeric data types

```python
df['Longitude'] = pd.to_numeric(df['Longitude'])
df['Latitude'] = pd.to_numeric(df['Latitude'])
```

Create a scatter map

```python
fig = px.scatter_mapbox(
    # Data frame and longitude and latitude columns
    data_frame=df,
    lat="Latitude",
    lon="Longitude",
    # Hover text configuration
    hover_name="Address",
    hover_data=["stationId"],
    # Marker settings
    size_max=30,
    opacity=0.4,
    # Plot configuration
    center={'lat':34.686567, 'lon':135.52000},
    zoom=4,
    height=600,
    width=800)

fig.update_layout(mapbox_style='open-street-map')
fig.update_layout(margin={"r": 0, "t": 0, "l": 0, "b": 0})
fig.show()
```
